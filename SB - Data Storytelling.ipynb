```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/skim1029/Downloads/starbucks_drinkMenu_expanded.xlsx')
```

```python
df
```

```python
#Count of each type of beverage on the starbucks menu
df_count = df.groupby(['Beverage_category'],sort=False)['Beverage'].nunique().reset_index()
df_count
```

```python
df_bar = df.sort_values(['Calories'],ascending=False)
df_bar
```

```python
df_bar = df_bar[['Beverage','Calories']].plot(kind='bar', title ="Calories per Beverage", figsize=(35, 25), legend=True, fontsize=12)
df_bar.set_xlabel("Beverage", fontsize=12)
df_bar.set_ylabel("Calories", fontsize=12)
plt.show()
```

```python
# the histogram of the data
df = df.sort_values(['Calories'],ascending=False)
df.hist(column='Calories', bins=10)
```

```python
df
```

```python
df.plot(x='Calories', y=['Protein','Sugars'])
plt.show()
```
